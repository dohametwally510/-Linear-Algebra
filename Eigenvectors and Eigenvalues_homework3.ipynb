```python
import numpy as np
```

```python
A = np.array([[0,1],
              [-2,-3]])

val ,vec =np.linalg.eig(A)
print("Eig Values: \n",val)
print("\nEig Vectors: \n",vec)
```

```python
D = np.diag(val)
print('B Diagonal Matrix\n',D)
```

```python
v =np.array([[1],
              [5]])
T = vec @ D**10 @ np.linalg.inv(vec) 
E_T= T @ v
print("Using Diagonalization:\n", E_T)
```

```python
M = np.array([[1,2,3,4],
             [5,5,6,7],
             [1,4,2,3],
             [5,3,2,1],
             [8,1,2,2]]) 
```

```python
#Step 1: Standardize the dataset:
mean=np.mean(M,axis=0)
std=np.std(M,axis=0)
S=(M-mean) /std
print("Standardized data:\n",S)
```

```python
#Step 2: Calculate the covariance matrix for the features in the dataset.
cov=(S.T @ S) / len(S)
print("Using the formula:\n",cov)
```

```python
#Step 3,4: 

val ,vec =np.linalg.eig(cov)
print("Eig Values: \n",val)
print("\nEig Vectors: \n",vec)
```

```python
#Step 5: Pick k eigenvalues and form a matrix of eigenvectors.
vec[ : ,0:2]
T = S @ vec[ : ,0:2]
np.around(T,3)
print("Transformed Data:\n",np.round(T,3))
```
