####  transformation type : Rotation

its matrix:

`[ X new]  =  [cos0  -sin0]   x   [X old]
 [ Y new]     [sin0   cos0]       [Y old]`


####  transformation type : Shearing

its matrix:

`[ X new]  =  [1  Shx]   x  [X old]
 [ Y new]     [0   1]       [Y old]`
